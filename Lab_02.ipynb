# **Reading the Data form CSV**

```python
dataset = pd.read_csv('data.set.csv')
dataset.describe()
```

**Review the Data as average lot size & How old is the newest home**

```python
import datetime
avg_lot_size = dataset['Landsize'].mean()  # Mean of the 'Landsize' column
avg_lot_size_rounded = round(avg_lot_size)  # Round to nearest integer

# Step 4: Find the newest home and calculate its age
current_year = datetime.datetime.now().year  # Get current year
newest_home_year = dataset['YearBuilt'].max()  # Find the maximum 'YearBuilt'
newest_home_age = current_year - newest_home_year
print(f"Average lot size (rounded): {avg_lot_size_rounded}")
print(f"Age of the newest home: {newest_home_age} years")
```

# **Building Model**

## Sklearn Library for Decision Tree as Regression, Classification or Plot tree

```python

from sklearn.tree import DecisionTreeRegressor
# Define model. Specify a number for random_state to ensure same results each run
dataset_model = DecisionTreeRegressor(random_state=1)

# Fit model
dataset_model.fit(X, y)
```

# **Making predictions for the following 5 houses**

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(dataset_model.predict(X.head()))
```
