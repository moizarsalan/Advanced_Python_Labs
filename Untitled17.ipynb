# **Task 01: Applying Feature Scaling and Ridge Regression**
1. Select a new dataset related to a different real-world scenario (e.g., predicting house prices, car
sales, or energy consumption).
2. Preprocess the dataset by handling missing values, encoding categorical variables, and splitting the

4. Compare the results of both models in terms of Mean Squared Error (MSE) and R-squared values.

6. Plot the feature importance using model coefficients for both Linear and Ridge regression.



**Step 1: Load and Preprocess the Dataset**

First, let’s load the dataset and handle any missing values.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Social_Network_Ads.csv')

# Display the first few rows of the dataset
print(df.head())

# Check for missing values
print(df.isnull().sum())
```

**Step 2: Define Features and Target Variable**

For regression tasks, we’ll use Age and EstimatedSalary as features and Purchased as the target variable.

```python
# Define features and target variable for regression tasks
X_reg = df.drop('Purchased', axis=1)
y_reg = df['Purchased']
```

**Step 3: Split the Dataset**

Split the dataset into training and testing sets.

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets for regression tasks
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
```

**Step 4: Scale the Features**

Scale the features using StandardScaler.

```python
from sklearn.preprocessing import StandardScaler

# Scale the features for regression tasks
scaler_reg = StandardScaler()
X_train_reg_scaled = scaler_reg.fit_transform(X_train_reg)
X_test_reg_scaled = scaler_reg.transform(X_test_reg)
```

**Step 5: Apply Linear Regression**


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the Linear Regression model
linear_reg = LinearRegression()
linear_reg.fit(X_train_reg, y_train_reg)

# Predict on the test set using Linear Regression
y_pred_linear = linear_reg.predict(X_test_reg)

# Evaluate the Linear Regression model
mse_linear = mean_squared_error(y_test_reg, y_pred_linear)
r2_linear = r2_score(y_test_reg, y_pred_linear)

print(f'Linear Regression MSE: {mse_linear}')
print(f'Linear Regression R-squared: {r2_linear}')
```

**Step 6: Apply Ridge Regression**

```python
from sklearn.linear_model import Ridge

# Initialize and train the Ridge Regression model
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X_train_reg_scaled, y_train_reg)

# Predict on the test set using Ridge Regression
y_pred_ridge = ridge_reg.predict(X_test_reg_scaled)

# Evaluate the Ridge Regression model
mse_ridge = mean_squared_error(y_test_reg, y_pred_ridge)
r2_ridge = r2_score(y_test_reg, y_pred_ridge)

print(f'Ridge Regression MSE: {mse_ridge}')
print(f'Ridge Regression R-squared: {r2_ridge}')
```

**Step 7: Compare Feature Importance**

```python
import matplotlib.pyplot as plt

# Plot the coefficients for both models
plt.figure(figsize=(12, 6))

# Linear Regression coefficients
plt.subplot(1, 2, 1)
plt.bar(X_reg.columns, linear_reg.coef_)
plt.title('Linear Regression Coefficients')
plt.xticks(rotation=90)

# Ridge Regression coefficients
plt.subplot(1, 2, 2)
plt.bar(X_reg.columns, ridge_reg.coef_)
plt.title('Ridge Regression Coefficients')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

# **Task 2: Building a Logistic Regression Classifier**
1. Use a different dataset (e.g., customer churn, loan approval, or cancer detection) to build a Logistic
Regression model.
2. Perform data preprocessing, including feature scaling using StandardScaler.
3. Split the dataset into training and testing sets.
4. Train a logistic regression model on the training data and predict on the test data.
5. Evaluate the model using a confusion matrix, accuracy score, precision, recall, and F1-score.
6. Plot the ROC Curve and calculate the AUC (Area Under the Curve) for your model.
7. Visualize the results using a scatter plot for the training and testing sets, similar to the visualization
done in Lab Practice or using scatter plots

**Step 1: Define Features and Target Variable**

For classification tasks, we’ll use the same features and target variable.

```python
# Define features and target variable for classification task
X_clf = df.drop('Purchased', axis=1)
y_clf = df['Purchased']
```

**Step 2: Split the Dataset**

Split the dataset into training and testing sets.

```python
# Split the dataset into training and testing sets for classification task
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)
```

**Step 3: Scale the Features**

Scale the features using StandardScaler.

```python
# Scale the features for classification task
scaler_clf = StandardScaler()
X_train_clf_scaled = scaler_clf.fit_transform(X_train_clf)
X_test_clf_scaled = scaler_clf.transform(X_test_clf)
```

**Step 4: Train Logistic Regression Model**


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Initialize and train the Logistic Regression model
log_reg = LogisticRegression()
log_reg.fit(X_train_clf_scaled, y_train_clf)

# Predict on the test set using Logistic Regression
y_pred_log = log_reg.predict(X_test_clf_scaled)

# Evaluate the Logistic Regression model
conf_matrix = confusion_matrix(y_test_clf, y_pred_log)
accuracy = accuracy_score(y_test_clf, y_pred_log)
precision = precision_score(y_test_clf, y_pred_log)
recall = recall_score(y_test_clf, y_pred_log)
f1 = f1_score(y_test_clf, y_pred_log)

print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

**Step 5: ROC Curve and AUC**


```python
from sklearn.metrics import roc_curve, auc

# Calculate ROC curve and AUC for Logistic Regression model
y_prob_log = log_reg.predict_proba(X_test_clf_scaled)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test_clf, y_prob_log)
roc_auc = auc(fpr, tpr)

# Plot ROC curve for Logistic Regression model
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```
